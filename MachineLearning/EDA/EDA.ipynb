```python
ignore = ['Corona_result','Body_temperature','Age']

df = (df.set_index(ignore, append=True)
        .astype(bool)
        .reset_index(ignore)
       )
df = df.reindex(columns=['Age', 'Gender', 'Body_temperature', 'Dry_cough', 'Sore_throat',
       'Weakness', 'Breathing_problem', 'Drowsiness', 'Pain_in_chest',
       'Travel_history_to_infected_countries', 'Diabetes', 'Heart_disease',
       'Lung_disease', 'Stroke_or_reduced_immunity', 'Symptoms_progressed',
       'High_blood_pressue', 'Kidney_disease', 'Change_in_appetite',
       'Loss_of_sense_of_smell', 'Corona_result'])
df['Corona_result'] = df['Corona_result'].replace([0],'Mild')
df['Corona_result'] = df['Corona_result'].replace([1],'Moderate')
df['Corona_result'] = df['Corona_result'].replace([2],'Severe')
```
